### 10.Calculate interest <span style='color:red'>Astagfirula</span> 



```python
original_price=int(input('Original Amount :'))
increase_percentage=float(input('Enter Increase Rate in % :'))
total_duration=int(input('Enter No of Years :'))
no_of_periods=total_duration
increased_price=0

for each_period in range(no_of_periods):
    increased_price+=original_price*(increase_percentage/100)
#     current_price+=differnece_increase
total_price_now=original_price+increased_price
print(f'\nAfter {each_period+1} Years Complete,\nPrice Increased :{increased_price},\nTotal Price Now :{total_price_now}')
```

### 11.<span style='color:red'>Euclidean</span> distance



```python
from math import sqrt
first_coordinate=input('Enter  1st Coordiante Seperated By Comma ---> x,y <---- :')
second_coordinate=input('Enter  2nd Coordiante Seperated By Comma ---> x,y <---- :')

first_coord_ls=first_coordinate.split(',')
second_coord_ls=second_coordinate.split(',')

euclidean_dist = sqrt((int(first_coord_ls[0])-int(second_coord_ls[0]))**2 + (int(first_coord_ls[1])-int(second_coord_ls[1]))**2  )

print(f'\nfor coordinates {first_coordinate} and {second_coordinate} Euclidean distance : {euclidean_dist}')


```

### 19.Count <span style='color:red'>Alphabets</span>, <span style='color:red'>Numbers</span>,<span style='color:red'>Special Charcters</span>

```python
import re
sentence = input('Enter Sentence : ')
sentence=re.sub(r'[ ]','',sentence,flags=re.IGNORECASE)


total_len=len(sentence)
digit_chr_str=re.sub(r'[a-z]','',sentence,flags=re.IGNORECASE)
char_str=re.sub(r'\d','',digit_chr_str)

char_len=len(char_str)
digit_chr_len=len(digit_chr_str)

digit_len=digit_chr_len-char_len
letter_len=total_len-digit_chr_len

print('No of Letters : {0}\nTotal Digits : {1}\nTotal Special character  : {2}'.format(letter_len,digit_len,char_len))
```

### 20. <span style='color:red'> Pattern</span>

```python
x_max=int(input('Enter integer for maxium x-axis star :'))
y_max=abs(2*(1-x_max))+1 #control y-axis range using x-axis max
# print(x_max,y_max)

for y in range(1,y_max+1):
    if y <= x_max:
        print('*'*y)
    else:
        print('*'*(y_max+1-y))
    
    
```

### 21. <span style='color:red'> Pattern</span>

```python
x_max=int(input('Enter integer for maxium x-axis star :'))
y_max=abs(2*(1-x_max))+1 #control y-axis range using x-axis max
# print(x_max,y_max)

for y in range(1,y_max+1):
    if y <= x_max:
        for x in range(1,y+1):
            print(x,end=' ')
    else:
        for x in range(1,(y_max+1)-y+1):
            print(x,end=' ')
    
    print('')
```
